# detection

```python
MODEL_FILENAME1 = "D:/Work/Picture/multy/"+a+"/code/X_model.hdf5"
MODEL_LABELS_FILENAME1 = "D:/Work/Picture/multy/"+a+"/code/X_model_labels.dat"
with open(MODEL_LABELS_FILENAME1, "rb") as f:
    lb1 = pickle.load(f)

# Load the trained neural network
model1 = load_model(MODEL_FILENAME1)

MODEL_FILENAME2 = "D:/Work/Picture/multy/"+a+"/code/Y_model.hdf5"
MODEL_LABELS_FILENAME2 = "D:/Work/Picture/multy/"+a+"/code/Y_model_labels.dat"
with open(MODEL_LABELS_FILENAME2, "rb") as f:
    lb2 = pickle.load(f)

# Load the trained neural network
model2 = load_model(MODEL_FILENAME2)

MODEL_FILENAME3 = "D:/Work/Picture/multy/"+a+"/code/Z_model.hdf5"
MODEL_LABELS_FILENAME3 = "D:/Work/Picture/multy/"+a+"/code/Z_model_labels.dat"
with open(MODEL_LABELS_FILENAME3, "rb") as f:
    lb3 = pickle.load(f)

# Load the trained neural network
model3 = load_model(MODEL_FILENAME3)



outputFolder = "D:/Work/Picture/multy/" + a + "/CI_1"

# 自動建立目錄
if not os.path.exists(outputFolder):
    os.makedirs(outputFolder)

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 480)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 512)
outputCounter = 0
count=0


while (outputCounter < 10):
    predictions = []

    # get a frameq
    ret, frame = cap.read(0)
    frame = frame[41:441, 57:457]
    r = 1000.0 / frame.shape[1]
    dim = (1000, int(frame.shape[0] * r))
    frame = cv2.resize(frame, dim, interpolation=cv2.INTER_AREA)
    # show a frame

    cv2.imwrite("%s/PICTURE.jpeg" % (outputFolder), frame)
    time.sleep(0.05)
    test_img = np.array(
        Image.open(outputFolder + "/PICTURE.jpeg").resize([pixel, pixel]))
    test_img = np.expand_dims(test_img, axis=0)
    prediction1 = model1.predict(test_img)
    prediction2 = model2.predict(test_img)
    prediction3 = model3.predict(test_img)
    x = lb1.inverse_transform(prediction1)[0]
    y = lb2.inverse_transform(prediction2)[0]
    Z = lb3.inverse_transform(prediction3)[0]
    predictions.append(x)
    predictions.append(y)
    predictions.append(Z)
    Predicted_Probability1 = model1.predict(test_img / 255.0)
    c = max(Predicted_Probability1[0])
    Predicted_Probability2 = model2.predict(test_img / 255.0)
    d = max(Predicted_Probability2[0])

    x = str(x)[35:]
    y = str(y)[35:]
    z = str(Z)[35:]
    d_a = (1000 // (1000 - int(z) * 100) * tangle - tangle//2) // 2
    d_b = (1000 - 2 * d_a) // (mx+1)
    text = (x + ',' + y+ ',' + z)
    cv2.namedWindow('capture', cv2.WINDOW_NORMAL)
    cv2.rectangle(frame, (d_b * (mx - int(x)) + d_a, d_b * (my - int(y)) 
+ d_a),(d_b * (mx - int(x)) + tangle//2 + d_a, d_b * (my - int(y)) + tangle//2 + d_a),
                  (0, 255, 255), 3)
    
    cv2.putText(frame, text, (11 * (mx - int(x)) , 11 * (my - int(y)) ), cv2.FONT_HERSHEY_SIMPLEX,
                1, (0, 0, 255), 1, cv2.LINE_AA)
    cv2.imshow("capture", frame)

    
    localtime = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S_")
    ResultFolder = "D:/Work/Picture/result/"+a
    if not os.path.exists(ResultFolder):
        os.makedirs(ResultFolder)
    if count>=0:
        cv2.imwrite("D:/Work/Picture/Result/"+a+"/"+localtime+str(count)+".jpeg", frame)
        count+=1
    
    print(localtime)
    print(count)
    print(x, y, z)
    

    k = cv2.waitKey(30) & 0xff

    if k == 27:  # press 'ESC' to quit

        break
cap.release()
cv2.destroyAllWindows()
```
